```python
#Importing all the required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Loading Data Set and Exploring it
df=pd.read_csv(r"C:\Users\Dell\Downloads\Iris_data.csv")
df.head()
```

```python
df.describe()
```

```python
sns.pairplot(df, hue="species" )
```

```python
x=df[["sepal_length","sepal_width","petal_length","petal_width"]].values
y=df["species"]
```

```python
#Trainning and using Decision Tree Algo
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state = 0)
clf = DecisionTreeClassifier().fit(X_train, y_train)
pred_clf=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred_clf)
```

```python
#Visualising the Decision Tree
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(80,55.5))
plot_tree(clf,filled=True,rounded=True,feature_names=x)
```
